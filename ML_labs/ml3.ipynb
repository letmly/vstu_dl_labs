```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

prep_df = pd.read_csv('proced1.csv', encoding='cp1251')

prep_df.head()
```

```python
plt.hist(prep_df['income'], bins=30, color='black', alpha=0.8)
plt.xlabel('income')
plt.ylabel('spending_rating')
plt.title('spending_rating based on income')
plt.show()
```

```python
correlation = prep_df['age'].corr(prep_df['income'])
print("correlation between buyer age and income:", correlation)
```

```python
plt.boxplot(prep_df['income'])
plt.ylabel('income')
plt.title('Income emissions analysis')
plt.show()
```

```python
s_r = prep_df['spending_rating'].value_counts()
s_r.plot(kind='bar')
plt.xlabel('spending_rating')
plt.ylabel('id')
plt.title('spending_rating among buyers')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.subplot(2, 2, 1)
plt.hist(prep_df['income'], bins=30, color='black', alpha=0.8)
plt.xlabel('income')
plt.ylabel('spending_rating')
plt.title('spending_rating based on income')
plt.subplot(2, 2, 2)
plt.scatter(prep_df['age'], prep_df['income'], color='black')
plt.xlabel('age')
plt.ylabel('income')
plt.title('scatterplot between age and income')
plt.subplot(2, 2, 3)
plt.boxplot(prep_df['income'])
plt.ylabel('income')
plt.title('income emissions analysis')
plt.subplot(2, 2, 4)
sns.heatmap(prep_df.drop(['gender'], axis=1).corr(), annot=True, fmt='.2g' )
plt.title('heat map')
plt.tight_layout()
plt.show()
```

* 1. Вывести на экран первые 5 строк вашего пред обработанного датасета;

```python
prep_st_ds = pd.read_csv('study_performance.csv', encoding='cp1251')

prep_st_ds.head()
```

* 2. Проанализировать распределение переменных;
* 4. Исследовать вбросы и аномалии;

```python
for col in prep_st_ds.columns[-3:]:
    print(prep_st_ds[col].describe())
    plt.boxplot(prep_st_ds[col])
    plt.ylabel(col)
    plt.title(f'{col} analysis')
    plt.show()
```

* 3. Исследовать корреляцию между переменными;

```python
sns.heatmap(prep_st_ds.drop(prep_st_ds.columns[:-3], axis=1).corr(), annot=True, fmt='.2g' )
plt.title('heat map')
plt.tight_layout()
plt.show()
```

* 5. Исследовать категориальные переменные;


```python


for col in prep_st_ds.columns[:-3]:
    plt.figure(figsize=(8, 6))
    print(f"\n\nКатегориальная переменная {col}")
    print(prep_st_ds[col].describe())
    print(prep_st_ds[col].unique())
    plt.hist(prep_st_ds[col])
    plt.title(f'Распределение по {col}')
    plt.tight_layout()
    plt.show()
```

* 6. Визуализировать результат в общий дешборд.

```python
plt.figure(figsize=(12, 12))
plt.subplot(2, 2, 1)
plt.hist(prep_st_ds['reading_score'], bins=30, color='black', alpha=0.8)
plt.xlabel('reading_score')
plt.ylabel('parental_level_of_education')
plt.title('reading_score based on parental_level_of_education')

plt.subplot(2, 2, 2)
plt.scatter(prep_st_ds['reading_score'], prep_st_ds['writing_score'], color='black')
plt.xlabel('reading_score')
plt.ylabel('writing_score')
plt.title('scatterplot reading_score age and writing_score')

plt.subplot(2, 2, 3)
plt.boxplot(prep_st_ds['reading_score'])
plt.ylabel('reading_score')
plt.title('reading_score emissions analysis')

plt.subplot(2, 2, 4)
sns.heatmap(prep_st_ds.drop(prep_st_ds.columns[:-3], axis=1).corr(), annot=True, fmt='.2g' )
plt.title('heat map')
plt.tight_layout()
plt.show()
```
