* нормализация math_score

```python
import pandas as pd
from sklearn import preprocessing

print(prep_st_ds.info())
print("Средняя math_score до нормализации:",prep_st_ds[['math_score', 'reading_score', 'writing_score']].mean())
scaler = preprocessing.MinMaxScaler()
prep_st_ds[['math_score', 'reading_score', 'writing_score']] = scaler.fit_transform(prep_st_ds[['math_score', 'reading_score', 'writing_score']])


print("Средняя math_score после:", prep_st_ds[['math_score', 'reading_score', 'writing_score']].mean())
```

```python
from sklearn.preprocessing import LabelEncoder

prep_st_ds = pd.get_dummies(prep_st_ds, columns=['gender', 'race_ethnicity'])

# print(prep_st_ds.columns)

label_encoder = LabelEncoder()
for column in ['parental_level_of_education', 'lunch', 'test_preparation_course']:
    prep_st_ds[column] = label_encoder.fit_transform(prep_st_ds[column])

prep_st_ds.head()
```

```python
df_old = pd.read_csv('../csvs/study_performance.csv', encoding='cp1251')

fig, axes = plt.subplots(2, 3, figsize=(15, 10))

# Перебор столбцов для построения гистограмм до нормализации
for i, column in enumerate(['math_score', 'reading_score', 'writing_score']):
    sns.histplot(df_old[column], bins=10, kde=True, ax=axes[0, i])
    axes[0, i].set_title(f'Распределение {column} до нормализации')

for i, column in enumerate(['math_score', 'reading_score', 'writing_score']):
    sns.histplot(prep_st_ds[column], bins=10, kde=True, ax=axes[1, i])
    axes[1, i].set_title(f'Распределение {column} после нормализации')

plt.tight_layout()
plt.show()
```
