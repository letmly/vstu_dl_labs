```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

prep_df = pd.read_csv('../csvs/proced1.csv', encoding='cp1251')

prep_df.head()
```

```python
plt.hist(prep_df['income'], bins=30, color='black', alpha=0.8)
plt.xlabel('income')
plt.ylabel('spending_rating')
plt.title('spending_rating based on income')
plt.show()
```

```python
correlation = prep_df['age'].corr(prep_df['income'])
print("correlation between buyer age and income:", correlation)
```

```python
plt.boxplot(prep_df['income'])
plt.ylabel('income')
plt.title('Income emissions analysis')
plt.show()
```

```python
s_r = prep_df['spending_rating'].value_counts()
s_r.plot(kind='bar')
plt.xlabel('spending_rating')
plt.ylabel('id')
plt.title('spending_rating among buyers')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.subplot(2, 2, 1)
plt.hist(prep_df['income'], bins=30, color='black', alpha=0.8)
plt.xlabel('income')
plt.ylabel('spending_rating')
plt.title('spending_rating based on income')
plt.subplot(2, 2, 2)
plt.scatter(prep_df['age'], prep_df['income'], color='black')
plt.xlabel('age')
plt.ylabel('income')
plt.title('scatterplot between age and income')
plt.subplot(2, 2, 3)
plt.boxplot(prep_df['income'])
plt.ylabel('income')
plt.title('income emissions analysis')
plt.subplot(2, 2, 4)
sns.heatmap(prep_df.drop(['gender'], axis=1).corr(), annot=True, fmt='.2g' )
plt.title('heat map')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
from sklearn import preprocessing

print(prep_st_ds.info())
print(" math_score  :",prep_st_ds[['math_score', 'reading_score', 'writing_score']].mean())
scaler = preprocessing.MinMaxScaler()
prep_st_ds[['math_score', 'reading_score', 'writing_score']] = scaler.fit_transform(prep_st_ds[['math_score', 'reading_score', 'writing_score']])


print(" math_score :", prep_st_ds[['math_score', 'reading_score', 'writing_score']].mean())
```

```python
from sklearn.preprocessing import LabelEncoder

prep_st_ds = pd.get_dummies(prep_st_ds, columns=['gender', 'race_ethnicity'])

# print(prep_st_ds.columns)

label_encoder = LabelEncoder()
for column in ['parental_level_of_education', 'lunch', 'test_preparation_course']:
    prep_st_ds[column] = label_encoder.fit_transform(prep_st_ds[column])

prep_st_ds.head()
```

```python
df_old = pd.read_csv('../csvs/study_performance.csv', encoding='cp1251')

fig, axes = plt.subplots(2, 3, figsize=(15, 10))

#       
for i, column in enumerate(['math_score', 'reading_score', 'writing_score']):
    sns.histplot(df_old[column], bins=10, kde=True, ax=axes[0, i])
    axes[0, i].set_title(f' {column}  ')

for i, column in enumerate(['math_score', 'reading_score', 'writing_score']):
    sns.histplot(prep_st_ds[column], bins=10, kde=True, ax=axes[1, i])
    axes[1, i].set_title(f' {column}  ')

plt.tight_layout()
plt.show()
```
